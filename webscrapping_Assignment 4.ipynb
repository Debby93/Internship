### Question 1. Scrape the details of most viewed videos on YouTube from Wikipedia. 
Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A) Rank B) Name C) Artist D) Upload date E) Views

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos ')
```

```python
# creating variable list
Rank=[]
Name=[]
Artist=[]
Upload_Date=[]
Views=[]
```

```python
#scraping Ranks of the video
try:
    rank_tag=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr')
    for i in rank_tag:
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append('-')
#scraping names of the video
try:
    names=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
    for i in names:
        Name.append(i.text)
except NoSuchElementException:
    Name.append('-')
    
#scraping artist of the video
try:
    artists=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
    for i in artists:
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append('-')
    
#scraping dates of the video
try:
    date=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
    for i in date:    
        Upload_Date.append(i.text)
except NoSuchElementException:
    Upload_Date.append('-')
    
#scraping views of the video
try:
    views=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
    for i in views:
        Views.append(i.text)
except NoSuchElementException:
    Views.append('-')
```

```python
#Creating dataframe
Youtube=pd.DataFrame({"Ranks":Rank,"Names":Name,"Artists":Artist,"Uploaded Date":Upload_Date,"Views":Views})
Youtube
```

### 6.	Scrape the details of Highest selling novels.   
A)	Book name   
B)	Author name   
C)	Volumes sold   
D)	Publisher   
E)	Genre   
Url - https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare  

```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare')
```

```python
#creating list
Book_name=[]
Author_name=[]
Volumes_sold=[]
publisher=[]
genre=[]

for i in driver.find_elements(By.XPATH,'//table[@class="in-article sortable"]/tbody/tr/td[2]'):
    Book_name.append(i.text)
for i in driver.find_elements(By.XPATH,'//table[@class="in-article sortable"]/tbody/tr/td[3]'):
    Author_name.append(i.text)
for i in driver.find_elements(By.XPATH,'//table[@class="in-article sortable"]/tbody/tr/td[4]'):
    Volumes_sold.append(i.text)
for i in driver.find_elements(By.XPATH,'//table[@class="in-article sortable"]/tbody/tr/td[5]'):
    publisher.append(i.text)
for i in driver.find_elements(By.XPATH,'//table[@class="in-article sortable"]/tbody/tr/td[6]'):
    genre.append(i.text)
    
Novels=pd.DataFrame({"Book Name":Book_name,"Author Name":Author_name,"Volumes Sold":Volumes_sold,"Publisher":publisher,"Genre":genre})
Novels
```

```python
driver.close()
```
