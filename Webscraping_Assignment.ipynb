```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url= 'https://en.wikipedia.org/wiki/Main_Page'
page= requests.get(url)
soup= BeautifulSoup(page.text, 'html.parser')
headers=[]
for i in soup.find_all(['h1','h2','h3','h4','h5','h6','h7','h8']):
    headers.append(i.text)
print(headers)

df= pd.DataFrame(headers)
df
```

```python
#Question 6 6)	Write a python program to scrape the details of most downloaded articles from AI in last 90 days
page= requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
book=BeautifulSoup(page.content)

title= book.find('h2')
title
title=[]
for i in book.find_all('h2'):
    title.append(i.text)
title
```

```python
author= book.find('span', class_="sc-1w3fpd7-0 dnCnAO")
author
author=[]
for i in book.find_all('span', class_="sc-1w3fpd7-0 dnCnAO"):
    author.append(i.text)
author
```

```python
date=book.find('span',class_="sc-1thf9ly-2 dvggWt")
date
date=[]
for i in book.find_all('span',class_="sc-1thf9ly-2 dvggWt"):
    date.append(i.text)
date
```

```python
df=pd.DataFrame({"Title":title, "Author":author, "Published Date":date})
df
```
